```python
from datasets import load_dataset, Audio

dataset_name = "mozilla-foundation/common_voice_11_0"
language_code = "tr"

try:
    print(f"'{dataset_name}' ({language_code}) veri seti indirilmeye çalışılıyor...")
    common_voice_test = load_dataset(
        dataset_name,
        language_code,
        split="test",
        download_mode="force_redownload"
    )
    print("Test veri seti başarıyla yüklendi.")

    common_voice_test = common_voice_test.remove_columns([
        "accent", "age", "client_id", "down_votes", "gender",
        "locale", "path", "segment", "up_votes"
    ])
    common_voice_test = common_voice_test.cast_column("audio", Audio(sampling_rate=16000))
    print("Sütunlar kaldırıldı ve ses formatı ayarlandı.")
    print(common_voice_test)

except Exception as e:
    print(f"Veri seti yüklenirken bir hata oluştu: {e}")
```

```python

model_name_or_path = "aysunn/whisper-small-tr"

try:
    processor = WhisperProcessor.from_pretrained(model_name_or_path)
except Exception as e:
    print(f"'{model_name_or_path}' için işlemci yüklenemedi. Model adını kontrol edin veya modelinizle birlikte işlemciyi de (processor) push ettiğinizden emin olun.")
    print("Eğer işlemciyi modelle birlikte push etmediyseniz, orijinal modelin işlemcisini kullanabilirsiniz ancak dil ve görev ayarlarını kontrol etmeniz gerekebilir:")
    print("# processor = WhisperProcessor.from_pretrained('openai/whisper-small', language='Turkish', task='transcribe')")
    raise e

try:
    model = WhisperForConditionalGeneration.from_pretrained(model_name_or_path)
    model.config.forced_decoder_ids = None
except Exception as e:
    print(f"'{model_name_or_path}' için model yüklenemedi. Model adını kontrol edin.")
    raise e


device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print(f"Model ve işlemci '{model_name_or_path}' adresinden yüklendi ve '{device}' cihazına taşındı.")
```

```python
asr_pipeline = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    device=0 if device == "cuda" else -1
)

ISTENEN_ALT_KUME_BOYUTU = 800

if len(common_voice_test) >= ISTENEN_ALT_KUME_BOYUTU:
    test_edilecek_veri = common_voice_test.select(range(ISTENEN_ALT_KUME_BOYUTU))
    print(f"Hızlı test için {ISTENEN_ALT_KUME_BOYUTU} örneklik bir alt küme oluşturuldu.")
else:
    test_edilecek_veri = common_voice_test
    print(f"Test veri seti zaten küçük ({len(common_voice_test)} örnek), tamamı kullanılacak.")
all_predictions = []
all_references = []


num_examples = len(test_edilecek_veri)
batch_size = 8

print(f"\nTest veri seti üzerinde transkripsiyonlar üretiliyor (toplam {num_examples} örnek)...")
if hasattr(test_edilecek_veri, 'format'):
    print(f"test_edilecek_veri formatı: {test_edilecek_veri.format}")


```

```python
print(f"common_voice_test değişkeninin tipi: {type(common_voice_test)}")

if hasattr(common_voice_test, '__len__') and len(common_voice_test) > 0:
    num_to_check = min(batch_size, len(common_voice_test))
    print(f"\nİlk {num_to_check} örneğin tipleri kontrol ediliyor:")
    first_batch_slice = common_voice_test[0:num_to_check]
    print(f"Alınan dilimin (slice) tipi: {type(first_batch_slice)}")
    for idx, item_in_slice in enumerate(first_batch_slice):
        print(f"  Batch içindeki {idx}. elemanın tipi: {type(item_in_slice)}")
        if not isinstance(item_in_slice, dict):
            print(f"    UYARI: Bu eleman bir sözlük değil! İçerik: {item_in_slice}")
else:
    print("common_voice_test yüklenmemiş veya boş.")
```

```python
print(f"common_voice_test formatı: {common_voice_test.format}")
```

```python
print(f"\nTest veri seti üzerinde transkripsiyonlar üretiliyor (toplam {num_examples} örnek)...")
print(f"common_voice_test formatı: {common_voice_test.format}")

for i in range(0, num_examples, batch_size):

    current_batch_dict = common_voice_test[i : i + batch_size]

    if not isinstance(current_batch_dict, dict):
        print(f"UYARI: {i} indeksindeki batch bir sözlük değil, atlanıyor. Tipi: {type(current_batch_dict)}")
        continue

    audio_column_data = current_batch_dict.get("audio")
    sentence_column_data = current_batch_dict.get("sentence")

    if audio_column_data is None or sentence_column_data is None:
        print(f"UYARI: {i} indeksindeki batch'te 'audio' veya 'sentence' sütunu eksik, atlanıyor. Anahtarlar: {current_batch_dict.keys()}")
        continue

    if not (isinstance(audio_column_data, list) and isinstance(sentence_column_data, list)):
        print(f"UYARI: {i} indeksindeki batch'te sütunlar liste değil, atlanıyor.")
        continue

    current_batch_input_audio = []
    current_batch_references = []

    actual_items_in_this_batch = len(audio_column_data)

    for k in range(actual_items_in_this_batch):
        audio_sample_info = audio_column_data[k]
        sentence_text = sentence_column_data[k]
        if isinstance(audio_sample_info, dict) and "array" in audio_sample_info:
            current_batch_input_audio.append(audio_sample_info["array"])
            current_batch_references.append(sentence_text)
        else:
            print(f"  UYARI: Batch {i//batch_size}, içindeki {k}. örnek hatalı, atlanıyor. Ses bilgisi: {audio_sample_info}")

    if not current_batch_input_audio:
        print(f"Batch {i//batch_size} içinde işlenecek geçerli örnek bulunamadı, atlanıyor.")
        continue

    try:
        predictions = asr_pipeline(current_batch_input_audio, batch_size=len(current_batch_input_audio))

        predicted_texts = [pred["text"] for pred in predictions]

        all_predictions.extend(predicted_texts)
        all_references.extend(current_batch_references)

        processed_count_in_loop = i + len(current_batch_input_audio)
        print(f"{processed_count_in_loop} / {num_examples} örnek işlendi (Bu batch'te {len(current_batch_input_audio)} örnek).")

    except Exception as e:
        print(f"Batch {i//batch_size + 1} (indeks {i}) işlenirken ASR pipeline hatası oluştu: {e}")
        continue

print("\nTüm tahminler üretildi.")
```

```python
wer_metric = evaluate.load("wer")

if len(all_predictions) == len(all_references) and len(all_predictions) > 0 :
    wer = 100 * wer_metric.compute(predictions=all_predictions, references=all_references)
    print(f"\nTest Veri Seti Üzerindeki Kelime Hata Oranı (WER): {wer:.2f}%")

    print("\nÖrnek Tahminler ve Referanslar:")
    for j in range(min(10, len(all_predictions))):
        print(f"Referans : {all_references[j]}")
        print(f"Tahmin   : {all_predictions[j]}\n")
elif len(all_predictions) == 0 and len(all_references) == 0:
    print("\nHiçbir tahmin üretilemedi veya referans bulunamadı. WER hesaplanamıyor.")
else:
    print(f"Hata: Tahmin sayısı ({len(all_predictions)}) ile referans sayısı ({len(all_references)}) eşleşmiyor. Lütfen süreci kontrol edin.")
```
